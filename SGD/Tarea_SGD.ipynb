```python
#librer√≠as necesarias
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
```

```python
#dataset
df = pd.read_csv('ReporteFelicidad_2019.csv')
df.head()
```

```python
#selecci√≥n de caracter√≠sticas y target
X = df[[
    "GDP per capita",
    "Social support",
    "Healthy life expectancy",
    "Freedom to make life choices",
    "Generosity",
    "Perceptions of corruption"
]].values

y = df["Score"].values

#estandarizaci√≥n
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#modelo lineal
lr = LinearRegression()
lr.fit(X_scaled, y)

# Predicciones
y_pred = lr.predict(X_scaled)

# Evaluando
print("Pesos (coeficientes):", lr.coef_)
print("Sesgo (intercepto):", lr.intercept_)
print("MSE:", mean_squared_error(y, y_pred))
```

```python
#implementaci√≥n sgd
sgd_model = SGDRegressor(max_iter=1000, learning_rate='invscaling',random_state=11)
sgd_model.fit(X_scaled, y)

#reemplazando los pesos con los del modelo lineal
sgd_model.coef_ = lr.coef_.copy()
sgd_model.intercept_ = np.array([lr.intercept_])

#entrenando nuevamente para refinar con SGD los pesos
sgd_model.fit(X_scaled, y)

y_pred_sgd = sgd_model.predict(X_scaled)
mse_sgd = mean_squared_error(y, y_pred_sgd)

print("Modelo optimizado con SGD:")
print("Pesos:", sgd_model.coef_)
print("Sesgo:", sgd_model.intercept_)
print("MSE:", mse_sgd, "\n")
```
